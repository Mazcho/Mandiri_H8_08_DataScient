```python
temp = input("Ketikan temperaturyang ingin dikonversi, eg. 45F, 120 C: ")
degree = int(temp[:-1])
i_convertion = temp[-1]

if i_convertion =="C":
    result = int(round((9*degree)/5+32))
    print("Temperaturnya adlah ",result,"Derajat")
elif i_convertion == "F":
    result = int(round((degree -32)*5/9))
    print("Temperaturnya adlah ",result,"Derajat")
else:
    print("Masukan input yangbenar")
```
